```python
# Importando pandas para trabalhar com dataframes
import pandas as pd

# Importando matplotlib
import matplotlib

import numpy as np

# Importando a classe Arquivos  do arquivo arquivos.py
from arquivos import Arquivos

from pandas.plotting import scatter_matrix
```

```python
# Importacao da classe Arquivos
arquivos = Arquivos()

# Chamada do metodo para retornar o dataset
df = arquivos.retornar_dataset()

# Chamada do metodo para retornar as primeiras linhas do dataset
df.head()
```

```python
numericas_converter = [
    'Rooms',
    'Bedroom2',
    'Bathroom',
    'Car',
    'Propertycount',
]

arquivos.colunas
```

```python
df.info()
```

```python
# Avaliando os dados
df.describe().round(2)
```

```python
# Pegando somente a contagem
df.describe().round(2).reset_index().iloc[0,]
```

```python
# Preenchendo os valores faltantes
df.fillna(axis=0, method='ffill', inplace=True	)

# Verificando correlação entre os itens com a coluna Price
df.corr(numeric_only=True).loc["Price"].sort_values(ascending=False)
```

```python
# Correlacao entre todos os itens
df.corr(numeric_only=True)
```

```python
df.dtypes.reset_index()
```

```python
# Convertendo quarto para inteiro
for coluna in numericas_converter:
    df[coluna] = arquivos.converter_para_inteiro(df[coluna])


# Convertendo Codigo postal para String
df.Postcode = arquivos.converter_para_string(df, 'Postcode')

# Convertendo Date para data
df.Date = arquivos.converter_para_data(df.Date)
```
